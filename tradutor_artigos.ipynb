```python
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import os
from langchain_openai.chat_models.azure import AzureChatOpenAI

headers = {
    'User-Agent': 'Ariel Sartorio, https://github.com/arielsartorio',
    'From': 'sartoriodev@gmail.com'
}

# Carregar variáveis de ambiente do arquivo .env
load_dotenv()
AZURE_ENDPOINT = os.getenv("AZURE_ENDPOINT")
API_KEY = os.getenv("AZURE_OPENAI_KEY")

def text_from_url(url):
    page = requests.get(url, headers=headers)
    if page.status_code == 200:
        soup = BeautifulSoup(page.text, 'html.parser')
        text = soup.article.get_text()
        #limpando texto
        lines = (line.strip() for line in text.splitlines())
        clean_text = '\n'.join(part for part in lines if part)
        return clean_text
    else:
        print(f"Failed to fetch the URL. Status code: {page.status_code}")
        return None
```

```python
def translate_article(text, lang):
    client = AzureChatOpenAI(
        azure_endpoint = AZURE_ENDPOINT,
        api_key = API_KEY,
        api_version = '2024-02-15-preview',
        deployment_name = 'gpt-4o-mini',
        max_retries = 0
    )

    messages = [
        ('system', 'Você atua como tradutor de textos'),
        ('user', f'Traduza o {text} para o idioma {lang} e responda em markdown')
    ]

    response = client.invoke(messages)
    print(response.content)
    return response.content
```

```python
url = input('Digite a URL do artigo: ')
lang = input('Para qual idioma deseja traduzir: ')
text = text_from_url(url)
translate_article(text=text, lang=lang)
```
